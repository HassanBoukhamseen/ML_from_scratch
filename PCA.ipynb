```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PCA:
    def __init__(self, num_principal_comp):
        self.num_principal_comp = num_principal_comp

    def fit(self, X):
        self.mean = X.mean(axis=0)
        X -= self.mean
        covariance = np.cov(X.T)
        _, eigenvectors = np.linalg.eig(covariance)
        eigenvectors = np.sort(eigenvectors.T)[::-1]
        self.comp = eigenvectors[0 : self.num_principal_comp]

    def transform(self, X):
        X -= self.mean
        return X@self.comp.T

```
