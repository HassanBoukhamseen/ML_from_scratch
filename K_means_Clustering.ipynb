```python
import numpy as np
import numpy.linalg as la
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd 
```

```python
plt.style.use('classic')
X, y = datasets.make_blobs(n_samples=500, n_features=5, centers=5, cluster_std=1.4, random_state=111)
plt.scatter(X[:, 0], X[:, 1])
centroids = X[[0, 100, 200, 300, 400]]
num_iterations = 5
distances = [[] for i in range(centroids.shape[0])]
colors = ['green', 'red', 'orange', 'blue', 'brown']
for i in range(num_iterations):
    for j, center in enumerate(centroids):
        distances[j] = np.sqrt(np.sum((X - center)**2, axis=1))
    distances = np.array(distances)
    clusters = [[] for _ in range(centroids.shape[0])]
    for k, dist in enumerate(distances.T):
        cluster_idx = np.argmin(dist)
        clusters[cluster_idx].append(X[k])
    for q, cluster in enumerate(clusters):
        centroids[q] = np.mean(cluster, axis=0)
    clusters = np.array(clusters)
for i, cluster in enumerate(clusters):
    cluster = np.array(cluster).T
    plt.scatter(cluster[0], cluster[1], color=colors[i])
    plt.scatter(centroids[i][0], centroids[i][1],s = 130, marker = "x", color='black')
    
    
```
